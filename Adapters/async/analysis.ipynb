```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

with open('read_history.pkl', 'rb') as f:
    history = pickle.load(f)

start = history[0][0]
timestamps = [(h[0] - start) * 1e3 for h in history]
bytes = [h[1] for h in history]
indices = list(range(len(timestamps)))

plt.figure()
plt.grid()
plt.plot(timestamps, indices, '.-')
for t, i, b in zip(timestamps, indices, bytes):
    plt.annotate(b.decode('ASCII'), (t, i), ha='center', va='bottom')
```
